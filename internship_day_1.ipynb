```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/marketing_campaign.csv', sep='\t')
```

```python
df
```

```python
df.info()
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
```

```python
df.isna().sum().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for col in df.columns:
  if df[col].dtype == 'object':
    df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df
```

In this dataset
* there were 24 null values and we removed the null values.
* we converted df['Dt_Customer'] from object to datetime.
* there were no duplicates value in the dataset.
* we used label encoder to standardized the object data.
* we convertrd the columns header from upper case to lower case.
